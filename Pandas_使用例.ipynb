```python
# ===========================
# データをグルーピングする。
# ===========================
grp = df.groupby('ID_C', as_index=True) # as index=False で ID_C のインデックス化回避
df_grp = grp.mean()
df_grp = grp.sum()
df_grp = grp.max()
df_grp = grp.min()
df_grp = grp.var() # n-1法であることに注意!! numpy以外は通常はn-1法
df_grp = grp.std() # n-1法であることに注意!! numpy以外は通常はn-1法
df_grp = grp.size() # pandas series として格納される
df_grp = grp.describe() #n-1法であることに注意!! numpy以外は通常はn-1法
df_grp
```

```python
# ==========================================
# aggregate(集計という意味)でグルーピング
# ==========================================
df_grp = grp.agg(['count','mean'])  
df_grp = grp.agg( {'Value_A': 'mean' , 'Value_B': ['count','mean','min','max']} ) # ディクショナリで集計するカラムを指定を指定

# ちなみに np でcount を得るには ... np.count_nonzero()

# ■ aggを使えば、lambdaを使用することが可能！！！
df_grp = grp.agg(['count','mean',lambda x : np.std(x)])   # ※ np.std は 【n法】

# ■ aggを使えば、関数を使用することが可能！！！　■■■いい感じ■■■
def std_p(series):
    return np.std(series) 
df_grp = grp.agg({'Value_A':['count','mean',std_p]})

df_grp
```
